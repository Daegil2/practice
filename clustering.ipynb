```python
def plot_data(datasets, position, title):
  x,y=datasets
  plt.subplot(position)
  plt.title(title)
  plt.scatter(x[:,0],x[:,1])
```

```python
np.random.seed(0)
n_samples =1500
random_state= 0
noise =0.05 
```

```python
circles= datasets.make_circles(n_samples= n_sample,
                              factor=0.5,
                              noise =noise,
                              random_state=random_state)

moon = datasets.make_moons(n_samples=n_sample,
                           noise=noise,
                           random_state=random_state)

blobs =datasets.make_blobs(n_samples=n_sample,
                           random_state=random_state)

no_structure =np.random.rand(n_sample, 2), None
```

```python
plt.figure(figsize=(12,12))
plot_data(circles, 221, 'Circles')
plot_data(moon, 222, 'Moons')
plot_data(blobs, 223, 'Blobs')
plot_data(no_structure, 224, 'No_structure')
```

```python
def fit_predict_plot(model, dataset, position, title):
  x,y=dataset
  model.fit(x)
  labels =model.predict(x)
  colors =np.array(['#30A0DE','#E53A40','#090707' ])
  ax= plt.subplot(position)
  ax.set_title(title)
  ax.scatter(x[:,0],x[:,1], color=colors[labels])

```

```python
fig=plt.figure(figsize=(10,10))
fig.suptitle('K mean clustering algorithm')

fit_predict_plot(cluster.KMeans(n_clusters=2, random_state=random_state), circles, 221, 'Circles')
fit_predict_plot(cluster.KMeans(n_clusters=2, random_state=random_state), moon, 222, 'Moons')
fit_predict_plot(cluster.KMeans(n_clusters=2, random_state=random_state), blobs, 223, 'Blobs')
fit_predict_plot(cluster.KMeans(n_clusters=2, random_state=random_state), no_structure, 224, 'No strucure')
```

```python
def fit_predict_plot(model, dataset, position, title):
  x,y=dataset
  model.fit(x)
  labels =model.fit_predict(x)
  colors =np.array(['#30A0DE','#E53A40','#090707','#A59E21' ])
  ax= plt.subplot(position)
  ax.set_title(title)
  ax.scatter(x[:,0],x[:,1], color=colors[labels])
```

```python
model = cluster.AgglomerativeClustering(n_clusters=4, linkage='ward')
# ward: 분산을 가장 적게 증가시키는 두 클러스터를 병합한다
# average:  평균거리가 가장 짧은 두 클러스터를 병합
# complete: 최대 거리가 가장 짧은 두 클러스터를 병합
fig =plt.figure(figsize=(10,10))
fig.suptitle('AgglomerativeClustering')

fit_predict_plot(model, circles, 221, 'Circles')
fit_predict_plot(model, moon, 222, 'Moons')
fit_predict_plot(model, blobs, 223, 'Blobs')
fit_predict_plot(model, no_structure, 224, 'No strucure')

```
